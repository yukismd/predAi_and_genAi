```python
df_movie_count = df['movieId'].value_counts().reset_index()
#df_movie_count
df_movie_count_top = df_movie_count.iloc[:20, :]    # レビューの多さトップXのmovie
df_movie_count_top
```

```python
df_movie = pd.read_csv('tmp/ml-latest-small/movies.csv')
df_movie
```

```python
df_movie_sample = df_movie[df_movie['movieId'].isin(df_sample['movieId'].unique())].reset_index(drop=True)
df_movie_sample
```

```python
df_movie_sample.to_csv('tmp/data/movies_sample.csv', index=False)
```

```python
df_movie_sample['genres'] = [g.split('|') for g in df_movie_sample['genres']]
```

```python
df_movie_sample
```

```python
exploded_df = df_movie_sample.explode('genres')
exploded_df['count'] = 1
pivot_df = exploded_df.pivot_table(index='movieId', columns='genres', values='count', fill_value=0)
pivot_df = pivot_df.reset_index()
pivot_df
```

```python
pivot_df.columns
```

```python
pivot_df = pivot_df.rename(columns={s:'genre_{}'.format(s) for s in pivot_df.columns[1:]})
```

```python
df_movie_sample2 = pd.merge(df_movie_sample, pivot_df, on='movieId')
df_movie_sample2
```

```python
df_movie_sample2.describe()
```

```python
df_movie_sample2.to_csv('tmp/data/movies_sample_AllAttributes.csv', index=False)
```

```python
df_sample
```

```python
df_sample2 = pd.merge(df_sample, df_movie_sample2, on='movieId')
df_sample.shape, df_sample2.shape
```

```python
df_sample2.head()
```

```python
genres = df_movie_sample2.columns.to_list()[3:]
genres
```

```python
{g:'sum' for g in genres}
```

```python
df_user = df_sample2.groupby('userId').agg({g:'sum' for g in genres}).reset_index()
df_user
```

```python
df_user.describe()
```

```python
df_user['tmp_male'] = df_user['genre_Action'] + df_user['genre_Adventure'] + df_user['genre_Crime'] + df_user['genre_Sci-Fi'] + df_user['genre_War']
df_user['tmp_young'] = df_user['genre_Animation'] + df_user['genre_Children'] + df_user['genre_Adventure'] + df_user['genre_Fantasy']
```

```python
df_user[['tmp_male', 'tmp_young']]
```

```python
df_user[['tmp_male', 'tmp_young']].describe()
```

```python
df_user['tmp_male2'] = df_user['tmp_male'] + np.random.choice([-9, -4, 0, 0, 0, 0, 4, 9], df_user.shape[0])
df_user['tmp_young2'] = df_user['tmp_young'] + np.random.choice([-4, -2, 0, 0, 0, 0, 2, 4], df_user.shape[0])
```

```python
df_user.shape[0]
```

```python
df_user[['tmp_male', 'tmp_male2', 'tmp_young', 'tmp_young2']].describe()
```

```python
df_user['user_gender'] = pd.cut(df_user['tmp_male2'], bins=[-float('inf'), 11, float('inf')], labels=['F', 'M'])
df_user['user_generation'] = pd.cut(df_user['tmp_young2'], bins=[-float('inf'), 2,  6, float('inf')], labels=['Young', 'Middle', 'Senior'])
```

```python
df_user['user_gender'].value_counts()
```

```python
df_user['user_generation'].value_counts()
```

```python
#df_user[['user_gender', 'user_generation']].value_counts()
```

```python
pd.crosstab(df_user['user_gender'], df_user['user_generation'])
```

```python
df_user
```

```python
df_user2 = df_user[['userId','user_gender','user_generation']]
df_user2
```

```python
df_user.to_csv('tmp/data/user_sample_AllAttributes.csv', index=False)
df_user2.to_csv('tmp/data/user_sample.csv', index=False)
```
